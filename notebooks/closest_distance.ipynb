## Train

```python
import pandas as pd

# find the coordinates of train stations
train = pd.read_csv("../data/raw/metro_stations.csv")
train_x = []
train_y = []
train_list = train['the_geom'].to_list()
for i in train_list:
    new = i.split()
    train_x.append(new[1][1:])
    train_y.append(new[2][:-1])
train['Latitude'] = list(map(float, train_y))
train['Longitude'] = list(map(float, train_x))
df_train = train.loc[:, ['station', 'Latitude', 'Longitude']]
```

```python
# find the distance to the nearest train station
apa = close_loc(df_train, apa, "apartment", "train_station", API[0])
```

```python
house = close_loc(df_train, house, "house", "train_station", API[-1])
```

## Park

```python
# find the coordinates of parks
df_park = pd.read_csv("../data/raw/park_reserve.csv")
name = list(df_park["Facility Name"])
new = []
index = []

# remove duplicates
for i in name:
    if i not in new:
        new.append(i)
    else:
        idx = name.index(i)
        if idx not in index:
            index.append(idx)
df_park.drop(index)
df_park.dropna(subset=["Latitude"], inplace=True)
df_park.dropna(subset=["Longitude"], inplace=True)

# find the distance to the nearest park
apa = close_loc(df_park, apa, "apartment", "park", API[5])
house = close_loc(df_park, house, "house", "park", API[6])
```
