```python
import pandas as pd
```

```python
income = pd.read_csv("../data/raw/income.csv")
income.head()
```

```python
# extract victoria's data
income_new = income.iloc[584:1047]
```

```python
income_new.head()
```

```python
# clean noisy data
income_new.dropna(how='all', axis=1, inplace=True)
income_new.head()
```

```python
new = income_new.iloc[:, 0:7]
```

```python
# arrange data by SA2 name for future matching with domain data
new.set_axis(["SA2", "SA2 NAME", "2014-15", "2015-16", "2016-17", "2017-18", "2018-19"],axis = 1, inplace = True)
new.head()
```

```python
list_income = ["SA2 NAME", "2014-15", "2015-16", "2016-17", "2017-18", "2018-19"]
# recreate dataframe with only useful columns
df = new.loc[:, list_income]
df = df.rename(columns={'SA2 NAME':'suburb', '2014-15':'2014income', \
                                '2015-16':'2015income', '2016-17':'2016income', \
                                '2017-18':'2017income', '2018-19':'2018income'})
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.to_csv("../data/curated/income_person.csv")
```
