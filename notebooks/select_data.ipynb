```python
import seaborn as sns
import matplotlib.pyplot as plt


#boxplot-apartment-facilities

df1_a = pd.read_csv('../data/curated/match_clean_apartment.csv')
labels = 'school','train_station', 'fitness_gym', 'hospital', 'shopping_center', 'park','natural_increase_2016_17'

school = df1_a.school
train_station = df1_a.train_station
fitness_gym = df1_a.fitness_gym
hospital = df1_a.hospital
shopping_center = df1_a.shopping_center
park = df1_a.park
increase_16 = df1_a.natural_increase_2016_17

plt.grid(True) 
plt.boxplot([school, train_station, fitness_gym, hospital, shopping_center, park, increase_16],
            medianprops={'color': 'red', 'linewidth': '1.5'},
            meanline=True,
            showmeans=True,
            meanprops={'color': 'blue', 'ls': '--', 'linewidth': '1.5'},
            flierprops={"marker": "o", "markerfacecolor": "pink", "markersize": 5},
            labels=labels)
plt.yticks(np.arange(0, 1000, 50000))
plt.show()

```

```python
import seaborn as sns
from pandas import Series,DataFrame
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('../data/curated/match_house16_s.csv')
labels = 'school','train_station', 'fitness_gym', 'hospital', 'shopping_center', 'park','natural_increase_2016_17'
school = df.school
train_station = df.train_station
fitness_gym = df.fitness_gym
hospital = df.hospital
shopping_center = df.shopping_center
park = df.park
increase_16 = df.natural_increase_2016_17
```

```python

plt.figure(figsize=(15, 5))  # set the width and height

ax=plt.subplot(1,4,1)
sns.regplot(x='school', y='price', data=df, ax=ax,
            scatter_kws={'marker':'.','s':3,'alpha':0.3},
            line_kws={'color':'g'});

ax=plt.subplot(1,4,2)
sns.regplot(x='train_station', y='price', data=df, ax=ax,
            scatter_kws={'marker':'.','s':3,'alpha':0.3},
            line_kws={'color':'g'});

ax=plt.subplot(1,4,3)
sns.regplot(x='hospital', y='price', data=df, ax=ax,
            scatter_kws={'marker':'.','s':3,'alpha':0.3},
            line_kws={'color':'g'});

ax=plt.subplot(1,4,4)
sns.distplot(df['price'])

plt.show()

```
