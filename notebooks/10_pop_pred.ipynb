### Data cleaning


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../data/curated/population16-20.csv")
df.shape
```

```python
df.head(10)
```

```python
# check for missing value
df.isnull().sum()
```

```python
df.set_index(["suburb"], inplace=True)
df
```

```python
df = df.apply(pd.to_numeric)
```

```python
# check for unreasonable value
df = df[(df.erp_2016> 16) & (df.erp_2017> 16) & (df.erp_2018> 16) &(df.erp_2019> 16)
        &(df.erp_2020> 16)]
df.shape
```

```python
#seperate to 4 dfs
df_erp = df[['erp_2016','erp_2017','erp_2018','erp_2019','erp_2020']]
df_nc = df[['natural_increase_2016_17','natural_increase_2017_18','natural_increase_2018_19',
            'natural_increase_2019_20','natural_increase_2020_21']]
df_iter_mig = df[['net_internal_migration_2016_17','net_internal_migration_2017_18','net_internal_migration_2018_19',
            'net_internal_migration_2019_20','net_internal_migration_2020_21']]
df_os_mig = df[['net_overseas_migration_2016_17','net_overseas_migration_2017_18','net_overseas_migration_2018_19',
            'net_overseas_migration_2019_20','net_overseas_migration_2020_21']]
df_os_mig
```

### Model: population forecasting


```python
df_erp
```

```python
#for saving predicted erp
l21_erp = []
l22_erp= []
l23_erp= []
l24_erp = []
l25_erp = []
```

```python
# recreate dataframe with only year and population
for i in range(len(df_erp)):
    df1 = pd.DataFrame(data=[['2016',df_erp.iloc[i,0]],
                    ['2017',df_erp.iloc[i,1]],
                    ['2018',df_erp.iloc[i,2]],
                    ['2019', df_erp.iloc[i,3]],
                    ['2020',df_erp.iloc[i,4]]],
             columns = ['year','erp'],
             )
    y = df1['erp']
    x = df1['year']
    x = x.values.reshape(-1,1)
    y = y.values.reshape(-1,1)
    model = LinearRegression()
    model.fit(x, y)
    
    # predict populations for 3 years
    predict_y = model.predict(x)
    X2 = [[2021],[2022],[2023],[2024],[2025]]
    Y2 = model.predict(X2)
    l21_erp.append(Y2[0][0])
    l22_erp.append(Y2[1][0])
    l23_erp.append(Y2[2][0])
    l24_erp.append(Y2[3][0])
    l25_erp.append(Y2[4][0])
```

```python
len(l21_erp)
```

```python
import os
os.mkdir(r'../data/curated/predictions')
```

```python
# append to new dataframe
df_erp["erp_2021"] = l21_erp
df_erp["erp_2022"] = l22_erp
df_erp["erp_2023"] = l23_erp
df_erp["erp_2024"] = l24_erp
df_erp["erp_2025"] = l25_erp

df_erp.to_csv('../data/curated/predictions/erp.csv')
```
